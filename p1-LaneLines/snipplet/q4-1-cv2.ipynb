```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline

# Read in and grayscale the image
origin = mpimg.imread('test_images/test.jpg')

image = np.copy(origin)
imshape = image.shape
plt.figure()
plt.title("origin image")
plt.imshow(image)

image = mpimg.imread('test_images/test.jpg')
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)

ignore_mask_color = (255, 255, 255)
vertices = np.array([[(0,imshape[0]),(0, 0), (imshape[1], 0), (imshape[1],imshape[0])]], dtype=np.int32)
cv2.fillPoly(image, vertices, ignore_mask_color)
plt.figure()
plt.title("ignore_mask_color image (w, h) ")
plt.imshow(image)

image = np.copy(origin)
vertices = np.array([[(0,imshape[0]),(0, imshape[0]/2), (imshape[1], imshape[0]/2), (imshape[1],imshape[0])]], dtype=np.int32)
cv2.fillPoly(image, vertices, ignore_mask_color)
plt.figure()
plt.title("ignore_mask_color image (w, h/2)")
plt.imshow(image)
```
